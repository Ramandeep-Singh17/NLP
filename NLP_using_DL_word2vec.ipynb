```python
import gensim.downloader as api

# Download the Word2Vec model using gensim.downloader and get its local path
# This will handle fetching the correct file and placing it in the gensim data directory.
model_path = api.load("word2vec-google-news-300", return_path=True)

# Load the model from the downloaded path, retaining the original binary and limit parameters
model = KeyedVectors.load_word2vec_format(model_path, binary=True, limit=500000)
```

```python
model['cricket']
```

```python
model.most_similar('man')
```

```python
model.most_similar('cricket')
```

```python
model.most_similar('facebook')
```

```python
model.similarity('man','woman')
```

```python
model.similarity('man','PHP')
```

```python
model.doesnt_match(['PHP','java','monkey'])
```

```python
vec = model['king'] - model['man'] + model['woman']
model.most_similar([vec])
```

```python
vec = model['INR'] - model ['India'] + model['England']
model.most_similar([vec])
```
